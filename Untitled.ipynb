```python
def convert_area(data):
    if "-" in data:
        data = data.split("-")
        newdata = (eval(data[0])+eval(data[-1]))//2
        return int(newdata)
    else:
        newdata = ""
        for i in data:
            if i not in "0987654321":
                break
            else:
                newdata += i

        return int(newdata)



# ans = convert_area("2100.54 - 2856")
df_new["total_sqft"] = df_new["total_sqft"].apply(convert_area)
# print(ans)
# print(type(ans))

```

```python
x = df_new.iloc[:,:-1]
y = df_new.iloc[:,-1]
```

```python
### now model building part

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
scaler.var_
```

```python
scaler.mean_
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,precision_score
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(x_train,y_train)
y_pred = lr_model.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
reciduals = y_test-y_pred
sns.distplot(reciduals)
```

```python
rf_model = RandomForestRegressor(n_estimators=100)
rf_model.fit(x_train,y_train)
y_pred = rf_model.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
reciduals = y_test-y_pred
sns.distplot(reciduals)
plt.show()
```

```python
dt_model = DecisionTreeRegressor()
dt_model.fit(x_train,y_train)
y_pred = dt_model.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
reciduals = y_test-y_pred
sns.histplot(reciduals,kde=True)
```

```python
gd_model = GradientBoostingRegressor(n_estimators=100)
gd_model.fit(x_train,y_train)
y_pred = gd_model.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
reciduals = y_test-y_pred
sns.histplot(reciduals,kde=True)
```

```python
sv_model = SVR()
sv_model.fit(x_train,y_train)
y_pred = sv_model.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
reciduals = y_test-y_pred
sns.histplot(reciduals,kde=True)
```
